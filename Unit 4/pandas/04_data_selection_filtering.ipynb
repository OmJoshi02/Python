# Filtering Techniques

```python
df[df['IMDb'] > 6] #Data where the IMDb rating is greater than 6
```

```python
#I want the names of actors where their movies are greater than 6 ratings

df[df['IMDb'] > 6]['Actor']
```

```python
#I want the data in which IMDb rating is greater than 6 and year should be greater than 2018

df[(df['IMDb']>6) & (df['Year']>2018)] 
```

```python
#I want movies greater than 2019 using query() method
#query() always returns copy
df.query("Year >2019") 
```

```python
df.query("IMDb > 7 and Year > 2016")
```
