```python
import pandas as pd
```

```python
df = pd.read_csv("data_cleaning_sample.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum() #Columnwise null values counting using sum()
```

```python
df.dropna() #It tells that the data is completely available row wise
```

```python
df.dropna(axis=1) #It tells that data is completely available columnwise
```

```python
df.fillna(0)
```

```python
df["Age"].fillna(df["Age"].mean()) #Replaces NaN value with average value of all values
```

```python
df.fillna(method="bfill") #bfill(backward fill) fills last known value from bottom to NA field
```

```python
df.fillna(method="ffill") #ffill(forward fill) fills last known value from top to NA field
```

```python
df.duplicated() #True if the row is duplicated
```

```python
df.drop_duplicates() #Removes duplicate rows
```

```python
df.duplicated(subset=["Name", "Age"]) #Columnwise duplicated finding
```

```python
df["Name"].str.lower() #Name column will be duplicated
```

```python
df2 = df.dropna().copy()
```

```python
#Convert all ages to int

df2["Age"] = df2["Age"].astype(int)
df2
```

```python
df2["Age group"] = df2["Age"].apply(lambda x : "Adult" if x >= 25 else "Minor")
df2
```

```python
#map() takes dictionary as input and converts all values in tables into the new values

gender_map = {"M":"Male", "F":"Female", "O":"Other"}
df["Gender"] = df["Gender"].map(gender_map)
df
```

```python
#replace() changes only dedicated value that is provided by dictionary way
df["City"] = df["City"].replace({"Delhi":"Del", "Mumbai":"Mum"})
df
```
