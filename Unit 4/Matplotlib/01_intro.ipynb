```python
import matplotlib.pyplot as plt
```

```python
#Simple Plot graph
#plt.plot() by default plots line graph
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
#Graph using values and adding lables to x and y axis

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's runs")
plt.plot(years, runs)
```

```python
#Multiple Lines in one plot

kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
sharma = [0, 300, 800, 1200, 1500, 1800, 1900, 2100, 500, 800]
plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")
plt.plot(years, sharma, label = "Rohit Sharma")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
#Graph using different lines '[color][marker][linetype]'
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]


plt.plot(years, kohli, 'ro--', label = "Kohli") #red circle with dashed line
plt.plot(years, sehwag, 'g^:', label = "Sehwag") #green triangle with dotted line
plt.legend()
```

```python
plt.plot(years, kohli, color = "orange", linestyle = '-.',label = "Kohli") 
plt.plot(years, sehwag, color = "blue", linestyle = '--', label = "Sehwag") 
plt.legend()
```

```python
plt.plot(years, kohli, color = "orange", linestyle = '-.', linewidth= 3, label = "Kohli") #linewidth increases width
plt.plot(years, sehwag, 'g^:', label = "Sehwag") #green triangle with dotted line
plt.legend()
```

```python
#Add grid lines in background

plt.plot(years, kohli, color = "orange", linestyle = '-.', linewidth= 3, label = "Kohli") #linewidth increases width
plt.plot(years, sehwag, 'g^:', label = "Sehwag") #green triangle with dotted line
plt.grid(True)
plt.legend()
```
