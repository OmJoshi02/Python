```python
import matplotlib.pyplot as plt
```

```python
activities = ['studying', 'playing', 'gaming', 'sleeping']
time_spent = [3,2,2,5]
colors = ['blue', 'green', 'orange', 'purple']

plt.figure(figsize = (6,6))
plt.pie(time_spent, labels = activities, colors = colors, startangle = 90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
studying = [3,4,3,5,4,3,4]
playing = [2,2,1,1,2,3,2]
tv = [2,1,2,2,1,1,1]
sleeping = [5,5,6,5,6,5,5]

labels = ["studing", "playing", "TV", "sleeping"]
colors = ['#FF6952', '#FFF952', '#BD52FF', '#FF5252']
plt.figure(figsize = (6,6))
plt.stackplot(days, studying, playing, tv, sleeping, labels = labels, colors = colors, alpha = 0.5)
plt.legend(loc="upper left")
plt.title("Student weekly activity")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
```
