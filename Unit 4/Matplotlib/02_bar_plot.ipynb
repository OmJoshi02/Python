```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.title("Year vs Runs bar")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.show()
```

```python
#vertical bar chart

import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years)) #Index positions of years
width = 0.25

plt.bar(x - width, sachin, width = width, label = "Sachin") #x-width = sachin is before x(left of sachin)
plt.bar(x, sehwag, width = width, label = "Sehwag") # x = sehwag is on x (middle sehwag)
plt.bar(x + width, kohli, width = width, label = "Kohli") #x + width = kohli is after sehwag(right of sehwag)

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years) # Shows years instead of 0,1,2,3,4 caused by numpy array
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Horizontal Bar graph

players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs, color = "purple")
plt.xlabel("Runs")
plt.ylabel("Players")
plt.title("Horizontal bar graph of players")
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Virat", "Rohit"]
runs = [1200, 1500, 1800]

plt.bar(players, runs, color = "blue")

#Add values on the top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha="center") #plt.text(X_pos, Y_pos, str(text name), ha="center")

plt.xlabel("Players")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()
```
