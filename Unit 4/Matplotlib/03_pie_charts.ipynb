```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Rohit"]
runs = [10000,8000,12000,5000]

plt.pie(runs, labels = labels)
plt.show()
```

```python
#Add hex colors
colors = ['#FF6952', '#FFF952', '#BD52FF', '#FF5252']
plt.pie(runs, labels = labels, colors = colors)
plt.show()
```

```python
#Wedgeprops

plt.pie(
    runs,
    labels = labels,
    colors = colors,
    wedgeprops={'edgecolor' : 'black', 'linewidth' : 1, 'linestyle' : '--'}
)
plt.show()
```

```python
#Explode : it will detach the given piece of pie chart

explode = [0.1, 0, 0, 0]

plt.pie(runs, explode = explode, labels = labels, colors = colors)
plt.show()
```

```python
#Shadow
plt.pie(runs, explode = explode, labels = labels, colors = colors, shadow = True)
plt.show()
```

```python
#startangle = to rotate the chart

plt.pie(runs, explode = explode, labels = labels, colors = colors, shadow = True, startangle = 30)
plt.show()
```

```python
#autopct : to show percentage of the piece on chart

plt.pie(runs, explode = explode, labels = labels, colors = colors, shadow = True, startangle = 30, autopct = '%1.1f%%')
plt.show()
```
